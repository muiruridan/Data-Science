```python
#Importing the necessary libraries

import pandas as pd
import numpy as np

# Loading the dataset
df =pd.read_csv('titanic.csv')
df.head()
```

```python
# Checking for duplicates
df.duplicated()
```

```python
# Checking data information
df.info()
```

```python
# Checking the categorigal and numerical columns

# Categorical columns
categorical_col = [col for col in df.columns if df[col].dtype == 'object']
print('Categorical columns: ', categorical_col)

# Numerical columns
numerical_col = [col for col in df.columns if df[col].dtype != 'object']
print("Numerical columns: ", numerical_col)
```

```python
# Unique values

df[categorical_col].nunique()
```

```python
# Checking unique ticket values

df['Ticket'].unique()[:50]
```

```python
# Dropping name and ticket columns

df1 = df.drop(columns=['Name', 'Ticket'])
df1.shape
```

```python
# Checking for null values
round((df1.isnull().sum()/df1.shape[0])*100, 2)
```

```python
df1.isnull().sum()
```

```python
# Dropping the null values
df2 = df1.drop(columns='Cabin')
df2.dropna(subset=['Embarked'], axis=0, inplace=True)
df2.shape
```

#### Missing Values Imputation

```python
# Mean imputation is suitable when the data is normally distributed and has no extreme outliers.
# Median imputation is preferable when the data contains outliers or is skewed.

# Mean imputation
df3 = df2.fillna(df2.Age.mean())

# Checking the null values again
df3.isnull().sum()
```

#### Handling Outliers

```python
# We can check if there are outliers in our dataset using a boxplot

import matplotlib.pyplot as plt

plt.boxplot(df3['Age'], vert=False)
plt.ylabel('Variable')
plt.xlabel('Age')
plt.title('Box Plot')
plt.show()
```

```python
# Calculate summary statistics
mean = df3['Age'].mean()
std = df3['Age'].std()

# Calculate the lower and upper bounds
lower_bound = mean - std * 2
upper_bound = mean + std * 2

print('Lower bound: ', lower_bound)
print('Upper bound: ', upper_bound)
```

```python
# Dropping the outliers
df4 = df3[(df3['Age'] >= lower_bound) & (df3['Age'] <= upper_bound)]
```

#### Data Transformation

```python
# This involves converting the data from one format to another to make it more suitable for analysis.

# Techniques: normalization and scaling.

X = df3[['Pclass','Sex','Age', 'SibSp','Parch','Fare','Embarked']]
Y = df3['Survived']
```

#### Data Formatting

```python
# Scaling - Involves transforming the values of features to a specific range.

# Min-Max Scaling

from sklearn.preprocessing import MinMaxScaler

# initialising the MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

# Numerical columns
num_col_ = [col for col in X.columns if X[col].dtype != 'object']
x1 = X
# learning the statistical parameters for each of the data and transforming
x1[num_col_] = scaler.fit_transform(x1[num_col_])
x1.head()
 
```
